```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Square function")
plt.xlabel("x: Sueldos")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```
