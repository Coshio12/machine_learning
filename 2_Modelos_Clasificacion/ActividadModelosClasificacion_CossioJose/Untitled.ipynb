```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('diabetes-dataset.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
#Se crean las variables y y X para realizar las operaciones
#La variable de salida Y es la columna "Outcome"
#La variable X serian todas las demas columnas.
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
#Conversion a numpy array
X = np.array(X)
print(X)
print('==========')
y = np.array(y)
print(y)
```

```python
#Reshape de las variables
X = X.reshape(-1,1)
print(X)
print('==========')
y = y.reshape(-1,1)
print(y)
```

```python
#Segmentacion de los datos
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
#Escalando variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
#Entrenamiento con el modelo de Regresion Logistica
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[1,150,50,30,100,44.0,0.6,22]])))
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
