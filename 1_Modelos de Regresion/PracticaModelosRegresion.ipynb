```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Dummy_Data_HSS.csv')
dataset.head(10)
```

```python
dataset.info()
```

```python
#Contando los datos nulos
dataset.isnull().sum(axis = 0)
```

```python
#Al ver que hay datos nulos y no son muchos, una opcion es eliminarlos
dataset.dropna(inplace = True)
```

```python
#Comprobando los datos nulos
dataset.isnull().sum(axis = 0)
```

```python
#Elejimos los valores para cada variable
X = dataset.iloc[:, 0].values #Se elige la columna de TV
y = dataset.iloc[:, -1].values #Se elige la columna de Sales
#La razon, es debido a que existe una linea de codigo
#dataset.corr() que muestra la correlacion de las columnas
#De las cuales, las columnas elegidas son las que tienen mayor correlacion entre ambas
```

```python
X
```

```python
y
```

```python
#Al momento de querer realizar el entrenamiento, daba un error que se solucionaba con un reshape
#Debido a eso, las variables "X" y "y" se los convirtio a numpy.array para hacer correctamente el reshape
#Pudiendo asi, realizar el entrenamiento con este modelo
```

```python
X = np.array(X)
X
```

```python
X = X.reshape(-1,1)
X
```

```python
y = np.array(y)
y
```

```python
y = y.reshape(-1,1)
y
```

```python
#Segmentacion de las variables
from sklearn.model_selection import train_test_split
X_train_Simple, X_test_Simple, y_train_Simple, y_test_Simple = train_test_split(X,y,test_size=0.2, random_state= 1)
```

```python
#REGRESION LINEAL SIMPLE
#========================
# Entrenamiento
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train_Simple, y_train_Simple)
```

```python
#Prediccion
y_pred = regressor.predict(X_test_Simple)
y_pred #respuestas del estudiante
```

```python
#REGRESION LINEAL MULTIPLE
#========================
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Codificacion de datos categoricos
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
#Segmentacion de los datos
from sklearn.model_selection import train_test_split
X_train_mult, X_test_mult, y_train_mult, y_test_mult = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Entrenamiento
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train_mult, y_train_mult)
```

```python
y_pred = regressor.predict(X_test_mult)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test_mult.reshape(len(y_test_mult),1)),1))
```
