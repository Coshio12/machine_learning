```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, -1].values
dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state= 1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train) #aqui es cuando entrena
```

```python
#Tomar examen, hacer las prediccion, usa los patrones de aprendizaje
y_pred = regressor.predict(X_test)
y_pred #respuestas del estudiante
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salario vs Experiencia (Datos del "estudiante")')
plt.xlabel('Años de experiencia')
plt.ylabel('Años de experiencia')
plt.show()
```
