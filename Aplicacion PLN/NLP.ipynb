```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importacion de los datos y manejo del dataset
```

```python
dataset = pd.read_csv('amazonreviews.tsv', delimiter = '\t', quoting = 3)
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum(axis = 0)
```

```python
#Cambiamos los valores del dataset (pos, neg) a
#numeros en este caso pos = 1, neg = 0
```

```python
dataset['label'] = dataset['label'].replace({'pos': 1, 'neg': 0})
```

```python
dataset.info()
```

```python
dataset
```

```python
# Eliminar registros a partir de la fila 999
dataset = dataset.iloc[:1000, :]
```

```python
dataset
```

```python
#Limpieza del Texto
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 1000):
  review = re.sub('[^a-zA-Z]', ' ', dataset['review'][i])
  review = review.lower()
  review = review.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
print(corpus)
```

```python
#Creamos el modelo Bolsa de Palabras
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 500)
X = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, 0].values
```

```python
#Dividimos el dataset en variables de entrenamiento y testeo
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
#Escalado de variables
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Entrenamiento con el modelo de Bosques Aleatorios
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#Prediccion del clasificador
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Generando la Matriz de Confusion
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
